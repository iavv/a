```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
import os

# Upload the TIFF image
uploaded = files.upload()
filename = next(iter(uploaded))

# Read the image using OpenCV
image = cv2.imread(filename, cv2.IMREAD_UNCHANGED)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for display

# Display the original TIFF image
plt.figure(figsize=(6, 6))
plt.title('Original TIFF Image')
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# Save the image with lossless compression (PNG format)
png_filename = 'image_lossless.png'
cv2.imwrite(png_filename, image)

# Read the saved PNG image
image_png = cv2.imread(png_filename, cv2.IMREAD_UNCHANGED)
image_png_rgb = cv2.cvtColor(image_png, cv2.COLOR_BGR2RGB)  # Convert to RGB for display

# Display the PNG image
plt.figure(figsize=(6, 6))
plt.title('PNG Image')
plt.imshow(image_png_rgb)
plt.axis('off')
plt.show()
```

```python
# Save the image with lossy compression (JPEG format)
jpeg_filename = 'image_lossy.jpg'
cv2.imwrite(jpeg_filename, image, [int(cv2.IMWRITE_JPEG_QUALITY), 90])

# Read the saved JPEG image
image_jpeg = cv2.imread(jpeg_filename, cv2.IMREAD_UNCHANGED)
image_jpeg_rgb = cv2.cvtColor(image_jpeg, cv2.COLOR_BGR2RGB)  # Convert to RGB for display

# Display the JPEG image
plt.figure(figsize=(6, 6))
plt.title('JPEG Image')
plt.imshow(image_jpeg_rgb)
plt.axis('off')
plt.show()
```

```python
# Function to get file size
def get_file_size(file_path):
    return os.path.getsize(file_path) / 1024  # Size in KB

# Print the dimensions and file sizes of each image
print(f'Dimensions of TIFF Image: {image.shape}')
print(f'Dimensions of PNG Image: {image_png.shape}')
print(f'Dimensions of JPEG Image: {image_jpeg.shape}')

print(f'File Size of TIFF Image: {get_file_size(filename):.2f} KB')
print(f'File Size of PNG Image: {get_file_size(png_filename):.2f} KB')
print(f'File Size of JPEG Image: {get_file_size(jpeg_filename):.2f} KB')
```

```python
# Plot the images together for comparison
plt.figure(figsize=(18, 6))

# Plot original TIFF image
plt.subplot(1, 3, 1)
plt.title('Original TIFF Image')
plt.imshow(image_rgb)
plt.axis('off')

# Plot PNG image
plt.subplot(1, 3, 2)
plt.title('PNG Image')
plt.imshow(image_png_rgb)
plt.axis('off')

# Plot JPEG image
plt.subplot(1, 3, 3)
plt.title('JPEG Image')
plt.imshow(image_jpeg_rgb)
plt.axis('off')

plt.show()
```
