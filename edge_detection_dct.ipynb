```python
import cv2
import matplotlib.pyplot as plt

# Step 1: Read the image
image = cv2.imread('/content/blackhite.jpg')
image1= cv2.imread('/content/face.jpg')

# Step 2: Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

# Step 3: Apply Gaussian Blur
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
blurred1 = cv2.GaussianBlur(gray1, (5, 5), 0)

# Step 4: Use Canny edge detection
edges = cv2.Canny(blurred, 50, 150)
edges1 = cv2.Canny(blurred1, 50, 150)

# Step 5: Display the original and edge-detected images
plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(122), plt.imshow(edges, cmap='gray'), plt.title('Edge Detected Image')
plt.show()

plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(122), plt.imshow(edges1, cmap='gray'), plt.title('Edge Detected Image')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def compute_dct(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Compute the DCT of the image
    dct_image = cv2.dct(np.float32(gray_image))

    return dct_image

# Read the images
image = cv2.imread('/content/blackhite.jpg')
image1= cv2.imread('/content/face.jpg')
# Compute the DCT of both images
dct_image = compute_dct(image)
dct_image1 = compute_dct(image1)

# Display the original images and their DCT
plt.figure(figsize=(10, 5))

# Original Image 1
plt.subplot(221)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image 1')
plt.axis('off')

# DCT Image 1
plt.subplot(222)
plt.imshow(np.log(np.abs(dct_image)), cmap='gray')
plt.title('DCT Image 1')
plt.axis('off')

# Original Image 2
plt.subplot(223)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Original Image 2')
plt.axis('off')

# DCT Image 2
plt.subplot(224)
plt.imshow(np.log(np.abs(dct_image1)), cmap='gray')
plt.title('DCT Image 2')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the pre-trained Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Read the images
image = cv2.imread('/content/blackhite.jpg')
image1= cv2.imread('/content/face.jpg')

# Convert the images to grayscale
gray1 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

# Detect faces in the images
faces1 = face_cascade.detectMultiScale(gray1, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
faces2 = face_cascade.detectMultiScale(gray2, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Draw rectangles around the detected faces
for (x, y, w, h) in faces1:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

for (x, y, w, h) in faces2:
    cv2.rectangle(image1, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Display the results
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Image 1 with Detected Faces')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Image 2 with Detected Faces')
plt.axis('off')

plt.show()

```
