```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Load the color image
image_path = '/perceptual-standard.jpg'
image = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Convert the image from BGR to RGB format for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#type of image\
print(f"RGB IMAGE")

# Display the original image size
print(f"Original RGB image size: {image.shape}")

# Define the new size
width, height = 900, 400

# Resize the image
resized_image = cv2.resize(image_rgb, (width, height))

# Display the resized image size
print(f"Resized RGB image size: {resized_image.shape}")

# Plot the original and resized images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axes[0].imshow(image_rgb)
axes[0].set_title(f'Original RGB Image\n{image.shape}')
axes[0].axis('off')

# Resized image
axes[1].imshow(resized_image)
axes[1].set_title(f'Resized  RGB Image\n{resized_image.shape}')
axes[1].axis('off')

# Display the plots
plt.show()
```

```python
# Load the grayscale image
image_path = '/gryascaleimg.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

#type of image\
print(f"GRAYSCALE IMAGE")

# Display the original image size
print(f"Original image: {image.shape}")

# Define the new size
width, height = 300, 200

# Resize the image
resized_image = cv2.resize(image, (width, height))

# Display the resized image size
print(f"Resized image: {resized_image.shape}")

# Plot the original and resized images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axes[0].imshow(image, cmap='gray')
axes[0].set_title(f'Original GrayScale Image\n{image.shape}')
axes[0].axis('off')

# Resized image
axes[1].imshow(resized_image, cmap='gray')
axes[1].set_title(f'Resized GrayScale Image\n{resized_image.shape}')
axes[1].axis('off')

# Display the plots
plt.show()
```

```python
# Load the color image
image_path = '/flower.jpg'  # Replace with the path to your image
image = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply a binary threshold to create a binary image
_, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

#type of image\
print(f"BINARY IMAGE")

# Display the original binary image size
print(f"Original binary image size: {binary_image.shape}")

# Define the new size
width, height = 300, 200  # Replace with your desired width and height

# Resize the binary image
resized_binary_image = cv2.resize(binary_image, (width, height))

# Display the resized binary image size
print(f"Resized binary image size: {resized_binary_image.shape}")

# Plot the original and resized binary images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Original binary image
axes[0].imshow(binary_image, cmap='gray')
axes[0].set_title(f'Original Binary Image\n{binary_image.shape}')
axes[0].axis('off')

# Resized binary image
axes[1].imshow(resized_binary_image, cmap='gray')
axes[1].set_title(f'Resized Binary Image\n{resized_binary_image.shape}')
axes[1].axis('off')

# Display the plots
plt.show()
```
