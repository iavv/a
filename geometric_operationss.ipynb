```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#create a black and white image
image = np.zeros((512,512,3), np.uint8)
image
```

```python
#draw line
cv2.line(image, (0,0), (0,0), (255,255,0), 5)
cv2_imshow(image)
```

```python
cv2.arrowedLine(image, (0,0), (511, 511), (255,255,0), 5)
cv2_imshow(image)
```

```python
cv2.rectangle(image, (295,0), (510,128), (0,255,0), 5)
cv2_imshow(image)
```

```python
cv2.circle(image, (447,63), 63, (0,0,255), -1)
cv2_imshow(image)
```

```python
cv2.putText(image, "Vaib", (10,500), cv2.FONT_HERSHEY_COMPLEX, 4, (255,255,255), 2)
cv2_imshow(image)
```
